```python
from bs4 import BeautifulSoup
import zipfile
from pymystem3 import Mystem
import string
```

```python
import pymorphy2
morph = pymorphy2.MorphAnalyzer()
```

```python
import spacy
nlp = spacy.load("ru_core_news_lg")
```

```python
m = Mystem()
```

```python
def get_extracted_text(name_of_file):
    soup = BeautifulSoup(name_of_file, "html.parser")
    text = soup.section
    return(text)    
```

```python
def get_plain_text(LibID, folder):
    archive = zipfile.ZipFile(folder, 'r')
    text = archive.read(str(LibID)+'.fb2')
    text = text.decode("windows-1251")
    text = get_extracted_text(text)
    text = str(text)
    return (text)
```

```python
def get_all_lemmas_pymorphy(plain_text):
    plain_text_without_punctuation = plain_text.translate(str.maketrans('', '', string.ascii_lowercase+string.punctuation))
    Our_List = plain_text_without_punctuation.split()
    my_lemma = ''
    for Word in Our_List:
        my_lemma = my_lemma + ' ' + (morph.parse(Word)[0].normal_form)
    return(my_lemma)
```

```python
def file_to_lemma_pymorphy(LibID, folder):
    example_text = get_plain_text(LibID, folder)
    return get_all_lemmas_pymorphy(example_text)
```

```python
def get_all_lemmas_mystem(plain_text):
    plain_text_without_punctuation = plain_text.translate(str.maketrans('', '', string.ascii_lowercase+string.punctuation))
    lemmas = m.lemmatize(plain_text_without_punctuation)
    my_lemma = ''.join(lemmas)
    return(my_lemma)
```

```python
def file_to_lemma_mystem(LibID, folder):
    example_text = get_plain_text(LibID, folder)
    return get_all_lemmas_mystem(example_text)
```

```python
def get_all_lemmas_spacy(plain_text):
    plain_text_without_punctuation = plain_text.translate(str.maketrans('', '', string.ascii_lowercase+string.punctuation))
    doc = nlp(plain_text_without_punctuation)
    my_lemma = " ".join([token.lemma_ for token in doc])
    return(my_lemma)
```

```python
def file_to_lemma_spacy(LibID, folder):
    example_text = get_plain_text(LibID, folder)
    return get_all_lemmas_spacy(example_text)
```
